```r
#aca mirar actividad por actividad, FILTRAR Y FIJAR CADA UNA, luego si clusters
#excel_sheets(ruta)
#Fermentacion energetica
FE <- (datas[(datas[,"Producto"]=="Fermentación entérica"), ])

FEm <- FE[,-c(1:7)]

#head(FEm)

### EUCLIDEAN ###
d_can3 = dist(FEm, method = "euclidean") 
s_can3=1/(1+d_can3)
s_mat3=as.matrix(s_can3)

jerarquico = hclust(d_can3, method = "average")

plot(jerarquico, labels = NULL, hang = 0.3, 
     main = "Dendrograma", sub = NULL,
     xlab = "Países", ylab = "Altura")

# Si definimos una altura (distancia) maxima entre clusters
#abline(h = 23, col = 'red')
rect.hclust(jerarquico , k = 4, border = 2:6)
```

```r
library(tidyverse)  # data manipulation
library(cluster)    # clustering algorithms
library(factoextra) # clustering algorithms & visualization

fviz_nbclust(as.matrix(d_can3), FUN = hcut, method = "silhouette") #The optimal number of clusters k is the one 
#that maximizes the average silhouette over a range of possible values for k. (2, 3, 4)

fviz_nbclust(as.matrix(d_can3), FUN = hcut, method = "wss") #(2,4)
```

```r
#reviso índice C (menos) y coeficiente CH (mas)
library(NbClust)
library(clusterCrit)

#Para 2 clusters
cat("2 clusters")
cc2s=hcut(FEm, k = 2, hc_method = "average", hc_metric = "euclidean")

intCriteria(as.matrix(FEm), cc2s$cluster, "C_index") 

intCriteria(as.matrix(FEm), cc2s$cluster, "Calinski_Harabasz")

#Para 4 clusters
cat("4 clusters")
cc2s=hcut(FEm, k = 4, hc_method = "average", hc_metric = "euclidean")

intCriteria(as.matrix(FEm), cc2s$cluster, "C_index") 

intCriteria(as.matrix(FEm), cc2s$cluster, "Calinski_Harabasz")

#dado que amba # de clusters tienen similar C, por CH es mejor 4 clusters
```

```r
sub_grp = cutree(jerarquico, k = 4) #revisar sub_grp#, man#
table(sub_grp)
fviz_cluster(list(data = d_can3, cluster = sub_grp), labelsize = 10)
```
